```python
from scipy import stats
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# This iris dataset comes bundled with sklearn
data = load_iris()
df = pd.DataFrame(data['data'], columns=data['feature_names'])
df['target'] = data['target']
```

```python
df.head()
```

```python
# Set the figure size and unpack the figure and axis tuple so you change attributes individually 
fig, ax = plt.subplots(figsize=(12, 8)) 
# Data from the dataset
ax.scatter(x = df['sepal length (cm)'], y = df['petal length (cm)'])

# Prevents the text at the top and adjusts teh subplot to fit into the area
plt.tight_layout()
```

```python
# Set the figure size and unpack the figure and axis tuple so you change attributes individually 
fig, ax = plt.subplots(figsize=(12, 8)) 

# Data from the dataset
ax.scatter(x = df['sepal length (cm)'], y = df['petal length (cm)'])

# Add a title to the Graph And Set It's Size 
ax.set_title("Sepal Length Vs. Petal Length in Irises", fontsize =30)

# Add Axis Title and make them easier to read
ax.set_ylabel('Sepal Length in CM', fontsize = 20)
ax.set_xlabel('Petal Length in CM', fontsize = 20)


#Change the size of the ticks
plt.xticks(fontsize=14)
plt.yticks(fontsize = 14)


# Prevents the text at the top and adjusts teh subplot to fit into the area
plt.tight_layout()
```

```python
# Set the figure size and unpack the figure and axis tuple so you change attributes individually 
fig, ax = plt.subplots(figsize=(12, 8)) 

# Data from the dataset
ax.scatter(x = df['sepal length (cm)'], y = df['petal length (cm)'], s = 100, c = df['target'])

# Add a title to the Graph And Set It's Size 
ax.set_title("Sepal Length Vs. Petal Length in Irises", fontsize =30)

# Add Axis Title and make them easier to read
ax.set_ylabel('Sepal Length in CM', fontsize = 20)
ax.set_xlabel('Petal Length in CM', fontsize = 20)


#Change the size of the ticks
plt.xticks(fontsize=14)
plt.yticks(fontsize = 14)

# Add the Grid and make it more transparant

ax.grid(which='major', alpha = .5)

# Prevents the text at the top and adjusts teh subplot to fit into the area
plt.tight_layout()
```
